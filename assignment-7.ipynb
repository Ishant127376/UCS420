```python
import numpy as np
array = np.array([[1, -2, 3],[-4, 5, -6]]) 
for i in range (array.shape[0]):
    for j in range (array.shape[1]):
        print(np.abs(array[i][j]),"\n")

fa = array.flatten()
per= np.percentile(fa,[25,50,75])
print( per)
per_col = np.percentile(array, [25, 50, 75], axis = 0)
print( per_col,"\n")
per_r = np.percentile(array, [25, 50, 75], axis = 1)
print( per_r,"\n")
mean_fa = np.mean(fa)
median_fa = np.median(fa)
std_fa = np.std(fa)
print(mean_fa,median_fa,std_fa,"\n")
mean_row = np.mean(array, axis = 1)
median_row = np.median(array, axis = 1)
std_row = np.std(array, axis = 1)
print(mean_row,median_row,std_row,"\n")
mean_col = np.mean(array, axis = 0)
median_col = np.median(array, axis = 0)
std_col = np.std(array, axis = 0)
print(mean_col,median_col,std_col,"\n")
```

```python
import numpy as np
a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])
floor_a = np.floor(a)
print(floor_a,"\n")
ceil_a = np.ceil(a)
print(ceil_a,"\n")
trunc_a= np.trunc(a)
print(trunc_a,"\n")
round_a = np.round(a)
print(round_a,"\n")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
rn = 102317060
np.random.seed(rn)

sales_data = np.random.randint(1000,5001,size = (12,4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns = categories, index = months)
print(df)
print(df.head())
print(df.describe())
df['Sum of Sales'] = df.sum(axis = 1)
category_sum = df[categories].sum(axis = 0)
print(category_sum,"\n")
print(df['Sum of Sales'],"\n")
growth_rate = df[categories].pct_change() * 100
print(growth_rate,"\n")
df['Growth Rate'] = df['Sum of Sales'].pct_change() * 100
print("DataFrame with Total Sales and Growth Rate added\n",df[['Sum of Sales', 'Growth Rate']], "\n")
df['Electronics'] *= 0.90
print(" 10% discount on Electronics category \n")
print("the new data frame is",df,"\n")

for i in categories:
    plt.plot(df.index,df[i])
plt.title('Sales trend on category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
sns.boxplot(data = df[categories])
plt.title('Category wise sales distribution')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```
