```python
import re
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
paragraph = """ Hi my name is Ishant Singh.
        I love to play bedminton and my favorite thing to do is
        writting songs and playing guitar. The best part about songs is that they heal you no matter how broken are you
        well they say some things can't be fixed but you  are not one of them alright that's me now tell me
        something about you. How are you?"""
text_lower = paragraph.lower()
clean_text = re.sub(r'[^\w\s]','',paragraph)
print(clean_text)
word_tokens = word_tokenize(clean_text)
sentence_tokens = sent_tokenize(clean_text)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in word_tokens if word_tokens if word not in stop_words]
freq_dist = FreqDist(filtered_words)
print('\n')
print("Word Tokenization", word_tokens)
print("Sentence Tokenization:", sentence_tokens)
print('\n')
print("Filtered Words - Stopwords removal-:", filtered_words)
print('\n')
fdist = FreqDist(filtered_words)
fdist.plot(10, title="Top 10 Word Frequencies (Excluding Stopwords)")
```

```python
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer

nltk.download('wordnet')

porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_porter = [porter.stem(word) for word in filtered_words]
stemmed_lancaster = [lancaster.stem(word) for word in filtered_words]
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Porter Stemmer:", stemmed_porter,'\n')
print("Lancaster Stemmer:", stemmed_lancaster,'\n')
print("Lemmatization:", lemmatized,'\n')
```

```python

import re

text =""" Hi my name is Ishant Singh.
        I love to play bedminton and my favorite thing to do is
        writting songs and playing guitar.I am 19 years old. The best part about songs is that they heal you no matter how broken are you
        well they say some things can't be fixed but you  are not one of them alright that's me now tell me
        something about you. How are you?"""

long_words = re.findall(r'\b\w{6,}\b', text)
print("Words with >5 letters:", long_words,'\n')

numbers = re.findall(r'\b\d+\b', text)
print("Numbers:", numbers,'\n')

capitalized = re.findall(r'\b[A-Z][a-z]+\b', text)
print("Capitalized words:", capitalized,'\n')

alpha_words = re.findall(r'\b[a-zA-Z]+\b', text)
print("Alphabetic words:", alpha_words,'\n')

vowel_words = re.findall(r'\b[aeiouAEIOU]\w+\b', text)
print("Words starting with vowel:", vowel_words,'\n')
```

```python

import re
def custom_tokenizer(text):
    tokens = re.findall(r"\b\w+(?:[-']\w+)*\b|\d+\.\d+|\d+", text)
    return tokens

sample_text = "Hi this is a sample text hello hello  testing 1 .. 2.. 3.. you there? hello?? this is test email test@email.com and a  test phone number 9876543210"
tokenized = custom_tokenizer(sample_text)
print("Custom tokens:", tokenized,'\n')

cleaned = re.sub(r'\S+@\S+', '<EMAIL>', sample_text)
cleaned = re.sub(r'https?://\S+', '<URL>', cleaned)
cleaned = re.sub(r'\+?\d{1,3}[\s-]?\d{10}|\b\d{10}\b', '<PHONE>', cleaned)
print("Cleaned text:", cleaned,'\n')
```
