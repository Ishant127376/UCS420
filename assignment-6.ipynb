```python
import numpy as np
import matplotlib.pyplot as plt
M = int(input("Enter a number"))
x = np.linspace(-10,10,100)
y1 =  M*x*x
y2 = M*np.sin(x)
plt.plot(x,y1,'b--',label='y=M*x^2')
plt.plot(x,y2,'r--',label='y=M*sin(x)')
plt.legend()
plt.grid()
plt.title("Mathematical Functions")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Cognitive Computing', 'Artificial Intelligence', 'Data Structures', 'Aptitude'],
    'Score': [85, 90, 78, 88, 95]
}
df = pd.DataFrame(data)
sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subject', y='Score', hue='Subject', data=df, palette="husl", legend=False)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title('Subject Scores', fontsize=14, fontweight='bold')
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.xticks(rotation=15)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102317060
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(np.cumsum(data), marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Cumulative Sum Line Plot")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Cumulative Sum")
axs[0, 0].grid(True)

x = np.arange(50)
noise = np.random.randn(50) * 0.5  # Adding some noise
axs[0, 1].scatter(x, data + noise, color='r', alpha=0.6)
axs[0, 1].set_title("Scatter Plot with Noise")
axs[0, 1].set_xlabel("Index")
axs[0, 1].set_ylabel("Value")
axs[0, 1].grid(True)

axs[1, 0].axis("off")
axs[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/company_sales_data.csv'
data = pd.read_csv(file_path)
print(data.head())

plt.figure(figsize=(10, 6))
sns.lineplot(x=data['month_number'], y=data['total_profit'], marker='o', color='b')
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in product_columns:
    sns.lineplot(x=data['month_number'], y=data[product], marker='o', label=product)
plt.title('Product Sales Data Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
data.set_index('month_number').plot(kind='bar', colormap='viridis', figsize=(12, 6))
plt.title('Bar Chart for All Features')
plt.xlabel('Month Number')
plt.ylabel('Sales / Profit')
plt.grid(True)
plt.show()
```
