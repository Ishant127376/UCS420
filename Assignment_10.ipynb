Question 1

```python
import re
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
paragraph = """ Hi my name is Ishant Singh.
        I love to play bedminton and my favorite thing to do is
        writting songs and playing guitar. The best part about songs is that they heal you no matter how broken are you
        well they say some things can't be fixed but you  are not one of them alright that's me now tell me
        something about you. How are you?"""
text_lower = paragraph.lower()
clean_text = re.sub(r'[^\w\s]','',paragraph)
print(clean_text)
word_tokens = word_tokenize(clean_text)
sentence_tokens = sent_tokenize(clean_text)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in word_tokens if word_tokens if word not in stop_words]
freq_dist = FreqDist(filtered_words)
print('\n')
print("Word Tokenization", word_tokens)
print("Sentence Tokenization:", sentence_tokens)
print('\n')
print("Filtered Words - Stopwords removal-:", filtered_words)
print('\n')
fdist = FreqDist(filtered_words)
fdist.plot(10, title="Top 10 Word Frequencies (Excluding Stopwords)")
```

Question 2

```python
alpha_words = re.findall(r'\b[a-zA-Z]+\b', paragraph)
print("Alphabetic words:", alpha_words)

filtered_alpha = [w for w in alpha_words if w.lower() not in stop_words]
print("Filtered:", filtered_alpha)

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered_alpha]
print("Stemmed:", stemmed)

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w.lower()) for w in filtered_alpha]
print("Lemmatized:", lemmatized)

print("Stemming may reduce words to non-readable roots (e.g., 'connects' → 'connect'), while Lemmatization gives valid words (e.g., 'devices' → 'device').")
```

Question 3

```python
texts = [
    "The iPhone 15 has a stunning display and improved battery life.",
    "Samsung Galaxy delivers top-tier performance and camera quality.",
    "Pixel phones offer pure Android experience with timely updates."
]

from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
bow = count_vect.fit_transform(texts)
print("Bag of Words (BoW):")
print(bow.toarray())
print("Feature Names:", count_vect.get_feature_names_out())

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(texts)
print("\nTF-IDF Scores:")
print(tfidf.toarray())
print("TF-IDF Feature Names:", tfidf_vect.get_feature_names_out())

import numpy as np
feature_names = tfidf_vect.get_feature_names_out()
for i, row in enumerate(tfidf.toarray()):
    top3_indices = np.argsort(row)[-3:][::-1]
    print(f"\nTop 3 keywords in text {i+1}:")
    for idx in top3_indices:
        print(f"{feature_names[idx]} (score: {row[idx]:.3f})")
```

Question 4

```python
text1 = """Artificial Intelligence is revolutionizing the tech industry.
It allows machines to mimic human intelligence and perform complex tasks."""
text2 = """Blockchain is a decentralized technology that secures data.
It ensures transparency and immutability in financial transactions."""

from nltk.tokenize import word_tokenize
text1_tokens = set(word_tokenize(text1.lower()))
text2_tokens = set(word_tokenize(text2.lower()))

intersection = text1_tokens.intersection(text2_tokens)
union = text1_tokens.union(text2_tokens)
jaccard_sim = len(intersection) / len(union)
print(f"Jaccard Similarity: {jaccard_sim:.3f}")

from sklearn.metrics.pairwise import cosine_similarity
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform([text1, text2])
cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print(f"Cosine Similarity: {cos_sim[0][0]:.3f}")

# 2c. Analysis
print("Cosine similarity captures the angle between vectors and often gives a better sense of content similarity, especially for longer texts.")
```

Question 6

```python
training_text = """Technology is advancing faster than ever. With artificial intelligence, we can automate tasks and create smart applications. The integration of IoT devices connects the physical world with the digital one. From smart homes to self-driving cars, innovation is shaping our future. As data grows, so does the need for powerful computing and secure networks."""

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np

tokenizer = Tokenizer()
tokenizer.fit_on_texts([training_text])
sequences = tokenizer.texts_to_sequences([training_text])[0]
vocab_size = len(tokenizer.word_index) + 1

input_sequences = []
for i in range(1, len(sequences)):
    n_gram_seq = sequences[:i+1]
    input_sequences.append(n_gram_seq)

max_len = max([len(seq) for seq in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')
X, y = input_sequences[:, :-1], input_sequences[:, -1]

model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=max_len-1))
model.add(LSTM(50))
model.add(Dense(vocab_size, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=200, verbose=0)

seed_text = "technology"
next_words = 5

for _ in range(next_words):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_len-1, padding='pre')
    predicted = model.predict(token_list, verbose=0).argmax()
    output_word = tokenizer.index_word.get(predicted, '')
    seed_text += " " + output_word

print("Generated text:", seed_text)
```
